```python
import pandas as pd 


data = {
    "Calories": [432, 567, 986],
    "Duration": [46, 78, 98],
}


df = pd.DataFrame(data)

print(df)
```

Locate rows

```python
print(df.loc[0])
```

Return a list of indexes

```python
print(df.loc[[0, 1]])
```

Named indexed

```python
df = pd.DataFrame(data, index = ["Day1", "Day2", "Day3"])


print(df)
```

Locating Named Indexes

```python
print(df.loc["Day2"])
```

Loading files into a dataframe

```python
df = pd.read_csv('data.csv')

print(df)
```
